```python
train_df = train_df.drop(columns=['Loan_ID'])
```

```python
tt_df = train_df[train_df['Credit_History'].isnull()]
tt_df
```

```python
tr_df = train_df.dropna()
tr_df
```

```python
xtrain = tr_df.drop('Credit_History',axis=1)
ytrain = tr_df['Credit_History']
```

```python
xtest = tt_df.drop('Credit_History',axis=1).dropna()
```

```python
xtraine = pd.get_dummies(xtrain)
ytraine = pd.get_dummies(ytrain, drop_first=True)
xteste = pd.get_dummies(xtest)
```

```python
logreg = LogisticRegression()
logreg.fit(xtraine,ytraine)
ypred = logreg.predict(xteste)
ypred
```
